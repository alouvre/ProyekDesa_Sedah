```python
# Membuat figure dan axes
fig, ax = plt.subplots()

# Mengatur warna untuk setiap kategori
warna_kategori = ['yellow', 'green', 'red']

# Membuat countplot menggunakan seaborn dan mengatur warna secara manual
sns.countplot(x='Kategori', data=Data_KrajanSidorejoJasem, palette=warna_kategori, ax=ax)

# Menambahkan judul
ax.set_title('Data Penduduk Desa Sedah')

# Menambahkan label persentase pada setiap bar
total_data = len(Data_KrajanSidorejoJasem['Kategori'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total_data)
    ax.annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Menampilkan plot
plt.show()
```
